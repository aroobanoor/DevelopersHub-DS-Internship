**Task 1: Exploring and Visualizing the Iris Dataset**
**Understanding how to read, summarize, and visualize a dataset using pandas, matplotlib, and seaborn.**

```python
# Importing the libraries we need
import pandas as pd           # for loading and working with data
import matplotlib.pyplot as plt  # for creating plots
import seaborn as sns          # for beautiful statistical charts
```

```python
# loading the iris dataset directly from seaborn
df = sns.load_dataset('iris')

# displaying the first 5 rows to see what the data looks like
print("First 5 rows of the dataset:")
df.head()
```

```python
# how many rows and columns
print("Shape of dataset (rows, columns):", df.shape)

# what are the names of column
print("\nColumn names:" , df.columns.tolist())

# statistics summary (mean, min, max, etc)
print("\nDataset Summary:")
df.describe()
```

```python
# checking to see if any values are missing
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# Scatter plot: relationship between sepal length and sepal width
# Each dot is colored by species (setosa, versicolor, virginica)
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='Set1')
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: relationship between sepal length and sepal width
# Each dot is colored by species (setosa, versicolor, virginica)
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='Set1')
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
# Box plot: detects outliers and shows spread of values per species
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='species', y='petal_length', palette='pastel')
plt.title('Petal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.tight_layout()
plt.show()
```

**Conclusion**
- The Iris dataset has 150 rows and 5 columns with no missing values.
- Scatter plots show that **setosa** is clearly separable from the other two species.
- Histograms reveal that petal measurements are more spread out than sepal measurements.
- Box plots confirm that **virginica** has the longest petals on average, and setosa has the smallest.
